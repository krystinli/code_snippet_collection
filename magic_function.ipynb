# Magic_Function
Specifal commands not built into python itself, always start with `%` <br />
Examples: 
- `%timeit`
- `%matplotlib inline`
- more [tut](https://medium.com/@rrfd/jupyter-notebook-shortcuts-tips-and-tricks-top-nbextensions-bring-order-to-your-notebooks-pt-636eb8ee5f89)

## 01_timeit
check the execution time of a python statement

```python
import numpy as np
import pandas as pd

a = np.random.randn(100,100)
%timeit np.dot(a, a)
```

This is similar to `time.time` function:

```python
import time
start_time = time.time()

# data loading with pandas 
df_chunk = pd.read_csv("dataset1/members.csv", chunksize=100000) 

chunk_list = []  
for chunk in df_chunk:  
    chunk_list.append(chunk)
data = pd.concat(chunk_list)

print("This process takes %s seconds" %(time.time()-start_time))
```
